### Scraping a website Using BeautifulSoup for Top 50 Artists in 2021

```python
import requests
from bs4 import BeautifulSoup

page = requests.get("https://top40weekly.com/top-50-artists-of-the-2021/").text
```

```python
soup = BeautifulSoup(page, "html.parser")
```

```python
artists = soup.find_all('li')
```

```python
# print(artists)
```

```python
attributes = []
for artist in artists:
    try:
        names = artist.contents[0]
        attributes.append(names.string)
    except IndexError:
        print("List Index is Out of Range for this HTML line, so Skipping this...")
```

```python
for i in attributes:
    print(i)
```

```python
attributes.index('Drake') # for first value in the top artists
```

```python
attributes.index('Wizkid') # for last value in the top artists
```

```python
top_50 = attributes[174:224].copy()
```

```python
print(top_50)
```

```python
print("Top 50 Artists of 2021")

for i in range(len(top_50)):
    print(i+1,'. '+top_50[i],sep = '')
```
